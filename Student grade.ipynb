```python
num_students =int(input("enter no.of students: "))
```

```python
students = {}

for s in range(num_students):
    print("\n--- Enter details for Student", s + 1, "---")
    student_id = input("Enter Student ID: ")
    name = input("Enter Student Name: ")

    subjects = ["Maths", "Science", "English", "History", "Computer"]
    marks = {}
```

```python
for subject in subjects:
        score = float(input(f"Enter marks for {subject}: "))
        marks[subject] = score
```

```python
total = sum(marks.values())
average = total / len(subjects)
grade = calculate_grade(average)
```

```python
# ✅ Store data for each student in the main dictionary
students[student_id] = {
        "Name": name,
        "Marks": marks,
        "Total": total,
        "Average": round(average, 2),
        "Grade": grade
}
```

```python
# ✅ Print all report cards after data entry
for student_id, info in students.items():
    print("\n===============================")
    print("        STUDENT REPORT CARD    ")
    print("===============================")
    print(f"Student ID   : {student_id}")
    print(f"Name         : {info['Name']}")
    print("-------------------------------")
    print("Subject-wise Marks:")
    for subject, score in info["Marks"].items():
        print(f"{subject:<12}: {score}")
    print("-------------------------------")
    print(f"Total Marks  : {info['Total']}")
    print(f"Average Marks: {info['Average']}")
    print(f"Grade        : {info['Grade']}")
    print("===============================")
```
